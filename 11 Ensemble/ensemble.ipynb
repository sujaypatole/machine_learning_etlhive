### Step 2: Seaparate X and Y(loan_status)

```python
X = df.drop(columns=["id","loan_status"])
Y = df[["loan_status"]]
```

```python
X.head()
```

```python
Y.head()
```

### Step 3 - Apply Preprocessing on X
cat - OneHot Encoding; con - StandardScaler

```python
X.dtypes
```

```python
cat = list(X.columns[X.dtypes == "object"])
con = list(X.columns[X.dtypes != "object"])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = make_pipeline(
   SimpleImputer(strategy="mean"),
   StandardScaler()
)
```

```python
cat_pipe = make_pipeline(
   SimpleImputer(strategy="most_frequent"),
   OneHotEncoder(handle_unknown="ignore", sparse_output=False)
)
```

```python
pre = ColumnTransformer(
   [
      ("num", num_pipe, con),
      ("cat", cat_pipe, cat)
   ]
).set_output(transform="pandas")
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

### Step 4 - Train Test Split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2, random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

### Step 5 - Build the model
1. Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model1 = RandomForestClassifier(
   n_estimators=1,
   min_samples_split = 2,
   min_samples_leaf=1,
   criterion="entropy"
)

model1.fit(xtrain, ytrain)
```

```python
model1.score(xtrain, ytrain)
```

```python
model1.score(xtest, ytest)
```

#### Step 5.1 Hyperparameter tunning

```python
params1 = {
   "n_estimators": [10,50,100,200],
   "max_depth": [3, 4, 5, 6]
}
```

```python
from sklearn.model_selection import RandomizedSearchCV

model_rf = RandomForestClassifier()

rscv_rf = RandomizedSearchCV(
   model_rf,
   param_distributions=params1,
   scoring="f1_macro",
   cv=5, n_jobs=-1
)

rscv_rf.fit(xtrain, ytrain)
```

```python
rscv_rf.best_params_
```

```python
rscv_rf.best_score_
```

```python
best_rf = rscv_rf.best_estimator_
best_rf
```

### Step 2: Model Evaluation for Random Forest

```python
best_rf.score(xtrain, ytrain)
```

```python
best_rf.score(xtest, ytest)
```

```python
ypred_test = best_rf.predict(xtest)
```

```python
ypred_test[0:5]
```

```python
ytest.head()
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_estimator(best_rf, xtest, ytest)
```

```python
from sklearn.metrics import classification_report

print(classification_report(ytest, ypred_test))
```

### Gradient Boosting Classifier
